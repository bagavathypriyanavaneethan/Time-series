**There are three missing values in the temp field within the specified range**

```python
a_df_imp=a_df['2015-02-21 10':'2015-02-21 23'][['TEMP']]
```

```python
a_df_imp
```

**Going to impute the missing values with the previous values**

# Forward filling

```python
a_df_imp['TEMP_FILL']=a_df_imp['TEMP'].fillna(method='ffill')
```

```python
a_df_imp
```

# Backward filling

```python
a_df_imp['Temp_bfill']=a_df_imp['TEMP'].fillna(method='bfill')
```

```python
a_df_imp
```

# Rolling window

```python
a_df_imp['Temp_roll']=a_df_imp['TEMP'].rolling(window=2,min_periods=1).mean()
```

```python
a_df_imp
```

**Imputing the data before 2 hours and 1 hour if not that data available**

# Imputing previous year data

```python
a_df.loc[a_df_imp.index+pd.offsets.DateOffset(years=-1)]['TEMP']
```

**Taking the data of same date and hour of the previous year**

```python
a_df_imp.index
```

```python
a_df_imp=a_df_imp.reset_index()
```

```python
a_df_imp['Temp_prev']=a_df_imp.apply(lambda x:a_df.loc[x['year_month_day_hour']-pd.offsets.DateOffset(years=-1)]['TEMP'] 
                                     if pd.isna(x['TEMP']) else x['TEMP'],axis=1)
```
